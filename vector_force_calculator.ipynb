```python
# Vector Force Calculator
# 2D and 3D resultant forces, magnitudes, direction angles

import numpy as np
import matplotlib.pyplot as plt

# define forces as vectors [Fx, Fy] in Newtons
# change these values to match any problem
F1 = np.array([3.0, 4.0])    # force 1
F2 = np.array([2.0, -1.0])   # force 2
F3 = np.array([-1.0, 2.0])   # force 3

# resultant force
R = F1 + F2 + F3

# magnitude
mag = np.linalg.norm(R)

# direction angle from positive x axis
angle = np.degrees(np.arctan2(R[1], R[0]))

print("=== 2D Vector Force Calculator ===")
print("F1: " + str(F1) + " N")
print("F2: " + str(F2) + " N")
print("F3: " + str(F3) + " N")
print("Resultant: " + str(R) + " N")
print("Magnitude: " + str(round(mag, 3)) + " N")
print("Direction: " + str(round(angle, 2)) + " degrees from x-axis")
```

```python
# 2D Force Diagram

fig, ax = plt.subplots(figsize=(8, 8))

# plot each force vector from origin
forces = [F1, F2, F3, R]
colors = ['steelblue', 'darkorange', 'green', 'red']
labels = ['F1', 'F2', 'F3', 'Resultant']

for F, color, label in zip(forces, colors, labels):
    ax.annotate('', xy=F, xytext=[0,0],
                arrowprops=dict(arrowstyle='->', color=color, lw=2))
    ax.text(F[0]*1.05, F[1]*1.05, label + ' ' + str(F) + ' N',
            color=color, fontsize=10)

ax.set_xlim(-3, 6)
ax.set_ylim(-3, 6)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('Fx (N)')
ax.set_ylabel('Fy (N)')
ax.set_title('2D Vector Force Diagram')
plt.tight_layout()
plt.savefig('vector_force_diagram.png')
plt.show()
print("diagram saved")
```

```python
# 3D Vector Force Calculator

# define 3D forces [Fx, Fy, Fz] in Newtons
F1_3d = np.array([2.0, 3.0, 1.0])
F2_3d = np.array([1.0, -2.0, 4.0])
F3_3d = np.array([-1.0, 1.0, 2.0])

# resultant
R_3d = F1_3d + F2_3d + F3_3d

# magnitude
mag_3d = np.linalg.norm(R_3d)

# unit vector
unit = R_3d / mag_3d

# direction cosines
cos_x = np.degrees(np.arccos(unit[0]))
cos_y = np.degrees(np.arccos(unit[1]))
cos_z = np.degrees(np.arccos(unit[2]))

print("=== 3D Vector Force Calculator ===")
print("F1: " + str(F1_3d) + " N")
print("F2: " + str(F2_3d) + " N")
print("F3: " + str(F3_3d) + " N")
print("Resultant: " + str(R_3d) + " N")
print("Magnitude: " + str(round(mag_3d, 3)) + " N")
print("Unit vector: " + str(np.round(unit, 3)))
print("Angle with x-axis: " + str(round(cos_x, 2)) + " degrees")
print("Angle with y-axis: " + str(round(cos_y, 2)) + " degrees")
print("Angle with z-axis: " + str(round(cos_z, 2)) + " degrees")
```
